```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from IPython.display import display  # Use display from IPython.display

# Base URL of the website
base_url = 'https://books.toscrape.com/catalogue/'

# URL of the page to scrape
url = base_url + 'page-1.html'

# Send an HTTP request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Initialize a dictionary to store the data
    books_data = {
        'Name': [],
        'Price': [],
        'Rating': [],
        'Link': []
    }

    # Find all the book articles
    book_articles = soup.find_all('article', class_='product_pod')

    # Iterate through each book article
    for book in book_articles:
        # Get the book title
        title_element = book.h3.a
        book_title = title_element.get('title')
        books_data['Name'].append(book_title)

        # Get the book price
        price_element = book.find('p', class_='price_color')
        book_price = price_element.text
        books_data['Price'].append(book_price)

        # Get the book rating
        rating_element = book.find('p', class_='star-rating')
        book_rating = rating_element.get('class')[1]  # The second class is the rating
        books_data['Rating'].append(book_rating)

        # Get the book link
        book_link = base_url + title_element.get('href')
        books_data['Link'].append(book_link)

    # Create a DataFrame with the data
    books_df = pd.DataFrame(books_data)

    # Display the DataFrame
    display(books_df)
else:
    print(f"Failed to retrieve the page. HTTP status code: {response.status_code}")
```
